```python
import nibabel as nib
import numpy as np
import scipy.ndimage
import matplotlib.pyplot as plt
import matplotlib as mpl
from scipy import stats

def initialiseLatexPlotting(fontSize=16):
    '''
        @summary: Make the plotting engine use latex, cause it looks much better.
    '''
    plt.rc( 'text', usetex=True )

   
    params = {'font.size' : fontSize,
              'text.latex.preamble': [ r"\usepackage{upgreek}",
                                       r"\usepackage{SIunits}"],
              }
    plt.rcParams.update(params)
```

```python
class compareHistoSegmentations:
    
    
    
    
    def __init__(self, imHighResName, imLowResName, sliceCorrespondenceDict):
        
        print("Reading images...")
        self.imH = nib.load( imHighResName )
        self.imL = nib.load( imLowResName  )
        
        self.imHDat = self.imH.get_data()
        self.imLDat = self.imL.get_data()
        
        self.spacingH = np.array( [ np.abs( self.imH.get_affine()[0,0] ), 
                                    np.abs( self.imH.get_affine()[1,1] ), 
                                    np.abs( self.imH.get_affine()[2,2] ) ] )
        
        print('Affine matrix high resolution image: ' )
        print self.imH.get_affine()
        print('Using spacing: ')
        print self.spacingH
        
        print self.imL.get_affine()
        
        self.spacingL = np.array( [ np.abs( self.imL.get_affine()[0,0] ), 
                                    np.abs( self.imL.get_affine()[1,1] ), 
                                    np.abs( self.imL.get_affine()[2,2] ) ] )

        print('Affine matrix low resolution image: ' )
        print self.imL.get_affine()
        print('Using spacing: ')
        print self.spacingL

        self.sliceCorrespondenceDict = sliceCorrespondenceDict
        #print('sliceCorrespondenceDict: ')
        #print sliceCorrespondenceDict




    def evaluate(self):
        
        print( "Starting evaluation of slice correspondences: ") 
        
        self.sliceWiseResults = []
        
        for hResSlice in self.sliceCorrespondenceDict.keys():
            lowResSlice = self.sliceCorrespondenceDict[hResSlice]
            print("HRes %i vs LRes %i " %(hResSlice, lowResSlice))
            self.sliceWiseResults.append( self._compareSlice(hResSlice, lowResSlice) ) 
        #
        # do the evaluation the other way around: Which volume error was made
        #
        curCoarseSlice = -1
        
        self.coarseSliceBasedResults = []
        
        for i, hResSlice in enumerate( self.sliceCorrespondenceDict.keys() ):
            
            # Check if the we need to start the next coarse level
            if curCoarseSlice != self.sliceCorrespondenceDict[hResSlice]:
                print( "Starting volume evaluation on next coarse level..." )

                curCoarseSlice = self.sliceCorrespondenceDict[hResSlice]
                
                self.coarseSliceBasedResults.append({})
                self.coarseSliceBasedResults[-1]['sumOverSegVol']  = 0.0
                self.coarseSliceBasedResults[-1]['sumUnderSegVol'] = 0.0
                self.coarseSliceBasedResults[-1]['sumVolumeH']     = 0.0
                self.coarseSliceBasedResults[-1]['sumVolumeL']     = 0.0
        
            self.coarseSliceBasedResults[-1]['sumOverSegVol']  += self.sliceWiseResults[i][ 'volOverSeg'  ]
            self.coarseSliceBasedResults[-1]['sumUnderSegVol'] += self.sliceWiseResults[i][ 'volUnderSeg' ]
            self.coarseSliceBasedResults[-1]['sumVolumeH']     += self.sliceWiseResults[i][ 'volL' ]
            self.coarseSliceBasedResults[-1]['sumVolumeL']     += self.sliceWiseResults[i][ 'volH' ]
    
        #
        # The global mean distance
        #
        globalAbsContourDistances = np.array([])
        meanAbsCountourDists = np.array([])
        rawValuesDistances = np.array([])
        globalVolOverSeg  = 0.0
        globalVolUnderSeg = 0.0
        globalVolH        = 0.0
        
        for i, hResSlice in enumerate( self.sliceCorrespondenceDict.keys() ):
            # self.sliceWiseResults[i][ 'volOverSeg'  ]
            
            globalAbsContourDistances = np.hstack( (globalAbsContourDistances, self.sliceWiseResults[i]['absContourDistances'] ) )
            #print globalAbsContourDistances
            #meanAbsCountourDists = np.hstack( (np.mean( globalAbsContourDistances ), self.sliceWiseResults[i]['absContourDistances']) ) 
            meanAbsCountourDists_1 = np.mean( globalAbsContourDistances )
            meanAbsCountourDists= np.append(meanAbsCountourDists, meanAbsCountourDists_1)
                
            
            globalVolOverSeg  += self.sliceWiseResults[i]['volOverSeg']
            globalVolUnderSeg += self.sliceWiseResults[i]['volUnderSeg']
            globalVolH        += self.sliceWiseResults[i]['volH']
        
        #print ('meanAbsCountourDists: ')
        #print meanAbsCountourDists
        
        
        for i, hResSlice in enumerate( self.sliceCorrespondenceDict.keys() ):
            # self.sliceWiseResults[i][ 'volOverSeg'  ]
            
            rawValuesDistances = self.sliceWiseResults[i]['absContourDistances']
            #print ('rawValuesDistances: ')
            #print rawValuesDistances
          
            #Write the array into a .txt. file in .csv format
            #np.savetxt("/Users/sarareis/Dropbox/PhD/SPIE/Manuscript/NewResults_December_interpolation/roi_2/ArrayMain{0}.csv" .format(i), rawValuesDistances, delimiter=",",fmt='%.7f')
        
        
        
        self.globalResults={ 'MeanAbsDistance' : np.mean( globalAbsContourDistances ), 
                             'StdAbsDistance'  : np.std( globalAbsContourDistances ),
                             'SemAbsDistance'  : stats.sem( globalAbsContourDistances ),
                             'VolumeOverSeg'   : globalVolOverSeg,
                             'VolumeUnderSeg'  : globalVolUnderSeg,
                             'VolumeH'         : globalVolH,
                             'meanAbsCountourDists': meanAbsCountourDists
                              }

    
    
    def plotResults(self, fontSize=12, plotDir = './'):
        
        def changeBoxPlotLook( bp ):
            # change the whisker look
            for w in bp['whiskers']:
                w.set_color('white')
                w.set_linestyle('-')
                ## change color and linewidth of the whiskers
                w.set(color='white', linewidth=2)
            
            for cap in bp['caps']:
                cap.set(color='white', linewidth=2)
            
            # change the color of the boxes
            for box in bp['boxes']:
                box.set_color='royalblue'
                box.set_facecolor('royalblue')
            
            # rehighlight the box medians
            for md in bp['medians']:
                md.set_color( 'red' )

   
        def applyPlotModifications( ax, printYLabel=True, printXLabel=True, setLowerAxLimit=True ):
            
            xLabels = [ '$0$', '', '', '', '', 
                        '$20$', '', '', '', '', 
                        '$40$', '', '', '', '',  
                        '$60$', '', '', '', '', 
                        '$80$', '', '', '', '', 
                        '$100$', '', '', '', '', 
                        '$120$']
            ax = plt.gca()
            
            if printXLabel:
                ax.set_xlabel( '$z\\;[\\unit{}{\\micro \\metre}]$' )
            
            if printYLabel:
                ax.set_ylabel( '$d \\;[\\unit{}{\\micro \\metre}]$' )
            
            ax.tick_params('x', pad=10)
            ax.tick_params('y', pad=10)
            ax.xaxis.labelpad=10
            #ax.set_aspect(0.375)
            ax.set_ylim(top    =  270.0)
            
            if setLowerAxLimit:
                ax.set_ylim(bottom = -100.0)
            
            ax.set_xticklabels( xLabels )
            ax.yaxis.grid( True, color='white', linestyle='-' )
            ax.spines['bottom'].set_color('white')
            ax.spines['top'].set_color('white')
            ax.spines['right'].set_color('white')
            ax.spines['left'].set_color('white')
            ax.xaxis.label.set_color('white')
            ax.yaxis.label.set_color('white')
            ax.tick_params(axis='x', colors='white')
            ax.tick_params(axis='y', colors='white')

            plt.subplots_adjust(bottom=0.16)
            plt.subplots_adjust(left=0.14)
        
        initialiseLatexPlotting( fontSize )

##        mpl.rcParams['text.usetex'] = True
##        mpl.rcParams['text.latex.preamble'] = [r'\usepackage{amsmath}']
        
        if not hasattr(self, 'sliceWiseResults'):
            print("Run evaluation first!")
            return
        
        zVals = np.mgrid[:self.sliceCorrespondenceDict.keys()[-1]+1] * self.spacingH[2] * 1000.0
        
        #
        # Get the evaluation results...
        #         
        diceVals    = []
        signedDists = []
        absDists    = []
        
        firstKey=self.sliceCorrespondenceDict.keys()[0]
        for i in range(firstKey):
            diceVals.append(np.nan)
            signedDists.append( np.array((0.0,0.0,0.0,0.0)) )
            absDists.append( np.array((0.0,0.0,0.0,0.0)) )
        
        for r in self.sliceWiseResults:
            
            # add a tiny fraction of noise to make the whiskers of the box plots end at the minimum/maximum...
            
            if np.percentile( r['signedContourDistances'], 25 ) == 0 or np.percentile( r['signedContourDistances'], 75 ) == 0:
                signedDists.append( r['signedContourDistances'] * 1000.0 + np.random.rand(r['signedContourDistances'].shape[0] ) * 0.005 * self.spacingH[0]  )  
            else:
                signedDists.append( r['signedContourDistances'] * 1000.0 )
            
            if np.percentile( r['absContourDistances'], 25 ) == 0 or np.percentile( r['absContourDistances'], 75 ) == 0:
                absDists.append( r['absContourDistances']  * 1000.0 + np.random.rand(r['absContourDistances'].shape[0] ) * 0.005 * self.spacingH[0] )  
            else:
                absDists.append( r['absContourDistances'] * 1000.0 )
            
            diceVals.append   ( r['Dice']                   )   
        #print absDists
        #Write the array into a .txt. file in .csv format
        #np.savetxt("/Users/sarareis/Dropbox/PhD/SPIE/Manuscript/NewResults_December_interpolation/roi_2/absDists.csv", absDists, delimiter=",")
        
             
        diceVals = np.array( diceVals )

        #
        # Plot the Dice coefficient
        #
        f=plt.gcf()
        f.clear()
        
        axes = f.add_subplot(1,1,1, axisbg='black')
        axes.plot( zVals, diceVals, 'D', color='royalblue' )
        #plt.plot( zVals, diceVals, 'D', color='royalblue' ) ##########
        plt.grid( color='grey', linestyle='-' )
        ax = plt.gca()
        
        ax.yaxis.grid( True, color='white', linestyle='-' )
        ax.spines['bottom'].set_color('white')
        ax.spines['top'].set_color('white')
        ax.spines['right'].set_color('white')
        ax.spines['left'].set_color('white')
        ax.xaxis.label.set_color('white')
        ax.yaxis.label.set_color('white')
        ax.tick_params(axis='x', colors='white')
        ax.tick_params(axis='y', colors='white')
        
        ax.set_ylim(top=1)
        ax.set_ylim(bottom=0.30)
        ax.set_xlim(left=0.0)
        ax.set_xlabel( '$z\\;[ \\unit{}{ \\micro \\metre } ]$' )
        ax.set_ylabel('$D(z)$')
        plt.savefig(plotDir + 'diceCoefficient.pdf', bbox_inches='tight', facecolor='black' )
        #plt.savefig(plotDir + 'diceCoefficient.png', bbox_inches='tight', facecolor='black' )
#         plt.show()
        #print(signedDists)
        
        #
        # Plot the signed distances wrt. the finest resolution
        #
        f=plt.gcf()
        f.clear()
        axes = f.add_subplot(1,1,1, axisbg='black')
        bp = axes.boxplot( signedDists, whis=np.inf, patch_artist=True )
        #bp = plt.boxplot( signedDists, whis=np.inf, patch_artist=True )
        changeBoxPlotLook(bp)
        ax = plt.gca()
        applyPlotModifications(ax, True, True, True)
        plt.savefig(plotDir + 'boxSignedContourDistance.pdf', bbox_inches='tight', facecolor='black' )
        #plt.savefig(plotDir + 'boxSignedContourDistance.png', bbox_inches='tight', facecolor='black' )
#         plt.show()

##        #
##        # Plot the asb distances wrt. the finest resolution
##        #
        f=plt.gcf()
        f.clear()
        axes = f.add_subplot(1, 1, 1, axisbg='black')
        bp = axes.boxplot( absDists, whis=np.inf, patch_artist=True )
#         bp=plt.boxplot( absDists, whis=np.inf, patch_artist=True )
        changeBoxPlotLook(bp) 
        ax = plt.gca()
        applyPlotModifications(ax, True, True, False)
        plt.savefig(plotDir + 'boxAbsContourDistance.pdf', bbox_inches='tight', facecolor='black' )
        #plt.savefig(plotDir + 'boxAbsContourDistance.png', bbox_inches='tight', facecolor='black' )
#         plt.show()
        
        #
        # Plot the coarse slice results
        #
        overSegs  = []
        underSegs = []
        volH      = []
        
        leftStart =[]
        width = 0.25
        dist  = 0.02
        
        for i in range( len( self.coarseSliceBasedResults ) ):
            
            overSegs.append ( self.coarseSliceBasedResults[i]['sumOverSegVol'] )
            underSegs.append( self.coarseSliceBasedResults[i]['sumUnderSegVol'] )
            volH.append     ( self.coarseSliceBasedResults[i]['sumVolumeH'] )
            leftStart.append (i+1)
        
        overSegs  = np.array( overSegs  )     
        underSegs = np.array( underSegs )
        volH      = np.array( volH )    
        leftStart = np.array( leftStart )
        
        f=plt.gcf()
        f.clear()

        plt.bar( leftStart,            overSegs,  width, color='b' )
        plt.bar( leftStart+width+dist, underSegs, width, color='r' )
        
        plt.savefig( plotDir + 'overUnderSegBar.pdf', bbox_inches='tight' )
        #plt.savefig( plotDir + 'overUnderSegBar.png', bbox_inches='tight' )
#         plt.show()
        
        #
        # Plot the relative over/under segmentation
        #
        f=plt.gcf()
        f.clear()

        plt.bar( leftStart,            overSegs  / volH * 100.0,  width, color='b' )
        plt.bar( leftStart+width+dist, underSegs / volH * 100.0, width, color='r' )
        ax = plt.gca()
        ax.set_ylabel('$V_{o/u}/V_H \\; [ \\% ]$')

        plt.savefig( plotDir + 'relativeOverUnderSegBar.pdf', bbox_inches='tight' )
        #plt.savefig( plotDir + 'relativeOverUnderSegBar.png', bbox_inches='tight' )
#         plt.show()
        
        
        
    def _compareSlice( self, highResSliceIndex, lowResSliceIndex ):
        
        #Create an array
        im2DH = np.array( self.imHDat[:,:,highResSliceIndex], dtype=np.int ) #high resolution stack
        im2DL = np.array( self.imLDat[:,:,lowResSliceIndex ], dtype=np.int ) #low resolution stack
        
        #Sum of array elements over a given axis
        magH = np.sum( im2DH )
        magL = np.sum( im2DL )
        
        intersectHL = np.sum( np.bitwise_and( im2DH, im2DL ) ) #Compute the bit-wise AND of two arrays element-wise.
        
        #Calculate the dice coefficient
        dice = 2.0*( intersectHL ) / ( magH + magL ) #formula: 2*(A intersect B)/(A+B)
        
        # Calculate the contour
        imgContourL = self._getCountour( im2DL )
        imgContourH = self._getCountour( im2DH )

        # Subtract the contour to get a distance of 0 on the contour. 
        distanceTransformHIn  = scipy.ndimage.distance_transform_edt(     im2DH-imgContourH, [self.spacingH[0], self.spacingH[1]], return_distances=True, return_indices=True )
        distanceTransformHOut = scipy.ndimage.distance_transform_edt( 1.0-im2DH, [self.spacingH[0], self.spacingH[1]], return_distances=True, return_indices=True )
        
        # combine into a signed distance function
        signedDistanceTransformH = distanceTransformHOut[0] - distanceTransformHIn[0]
        absDistanceTransformH    = distanceTransformHOut[0] + distanceTransformHIn[0]
        
        
        idxContourL = np.nonzero( imgContourL )
        
        # lookup the distances in the high-res image
        signedContourDistances = signedDistanceTransformH[ idxContourL ]
        absContourDistances    = absDistanceTransformH[ idxContourL ]
        
        #
        # Do some volume measurements
        #
        areaH = np.sum( im2DH )* self.spacingH[0] * self.spacingH[1]
        areaL = np.sum( im2DL )* self.spacingL[0] * self.spacingL[1]
        volH  = areaH * self.spacingH[2]
        volL  = areaL * self.spacingH[2]  # NOTE: The evaluation here is based on the finer slicing thickness!
        
        difHL = im2DH - im2DL
        
        # Region of under segmentation == positive values in difference image
        difHLPos = difHL.copy()
        difHLPos[ difHLPos < 0 ] = 0
        areaUnderSeg = np.abs( np.sum( difHLPos ) ) * self.spacingH[0] * self.spacingH[1]
        volUnderSeg  = areaUnderSeg * self.spacingH[2]
        
        # Region of over segmentation == negative values in difference image
        difHLNeg = difHL.copy()
        difHLNeg[ difHLNeg > 0 ] = 0
        areaOverSeg = np.abs( np.sum( difHLNeg ) ) * self.spacingH[0] * self.spacingH[1]
        volOverSeg = areaOverSeg * self.spacingH[2]
        
        resultDict = { 'Dice'                   : dice,
                       'absContourDistances'    : absContourDistances,
                       'signedContourDistances' : signedContourDistances,
                       'areaOverSeg'            : areaOverSeg, 
                       'volOverSeg'             : volOverSeg,
                       'areaUnderSeg'           : areaUnderSeg,
                       'volUnderSeg'            : volUnderSeg,
                       'volH'                   : volH,
                       'volL'                   : volL,
                       'areaH'                  : areaH ,
                       'areaL'                  : areaL 
                      }
                
        return resultDict
        
            
            
                 
    def _getCountour(self, im):
        
        erodedImg = scipy.ndimage.binary_erosion( im )
        return im - erodedImg




def printDistMeasures( prefix, comp ):
    
    numDistMeasures = []
    
    for i in range( len(comp.sliceWiseResults) ):
        numDistMeasures.append( comp.sliceWiseResults[i]['absContourDistances'].shape[0] )
        
    numDistMeasures = np.array( numDistMeasures )
    print( prefix )
    print( 'Mean (std) dist measures per slice: %.1f (%.1f)' %( np.mean(numDistMeasures), np.std(numDistMeasures) ) )
    
```
