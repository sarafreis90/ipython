```python
#THIS ONE - IT WOOOOORKS 
from Tkinter import Button
from Tkinter import *
import os, re, Tkinter, random
from PIL import Image
from PIL import ImageTk
import glob

# original code from: http://stackoverflow.com/questions/10599771/how-to-loop-through-subfolders-showing-jpg-in-tkinter
class SimpleAppTk(Tkinter.Frame):
    
    def __init__(self,*args,**kwargs):
        
        Tkinter.Frame.__init__(self,*args,**kwargs)
        #self.filelist = glob.glob("/Volumes/vph-prism5/Sara/observerStudy/test_images/*.png")  #get your files here
        self.filelist = glob.glob("/Users/sarareis/Documents/observerStudy/test_images/*.png")
        self.index=-1
        self.setup()
        self.classification = []
        self.display_next_m()
        self.display_next_i()
        

    def setup(self):
        self.Label=Tkinter.Label(self)
        self.Label.grid(row=0,column=0)
        #self.Button=Tkinter.Button(self,text="Next",command=self.display_next)
        #self.Button.grid(row=0,column=1)
        self.Button1 = Tkinter.Button(self,text="Mature", command = lambda: self.display_next_m())
        self.Button2 = Tkinter.Button(self,text="Immature", command = lambda: self.display_next_i())
        self.Button1.grid(row=0,column=1)
        self.Button2.grid(row=0,column=2)
    
    def display_next_m(self):
        self.index+=1
        try:
            f=self.filelist[self.index]
            print "You chose mature." 
            self.classification.append('mature')
            print self.classification
        except IndexError:
            self.index=-1  #go back to the beginning of the list.
            self.display_next_m()
            return
        
        #create PhotoImage here
        print f
        photoimage=Image.open(f)
        resized = photoimage.resize((int(photoimage.size[0]*.3),int(photoimage.size[1]*.2)),Image.ANTIALIAS)
        tkpi = ImageTk.PhotoImage(resized)
        self.Label.config(image=tkpi)
        self.Label.image=tkpi
    
    
    def display_next_i(self):
        self.index+=1
        try:
            f=self.filelist[self.index]
            print "You chose immature." 
            self.classification.append('immature')
            print self.classification
        except IndexError:
            self.index=-1  #go back to the beginning of the list.
            self.display_next_i()
            return
        
        #create PhotoImage here
        print f
        photoimage=Image.open(f)
        resized = photoimage.resize((int(photoimage.size[0]*.3),int(photoimage.size[1]*.2)),Image.ANTIALIAS)
        tkpi = ImageTk.PhotoImage(resized)
        self.Label.config(image=tkpi)
        self.Label.image=tkpi

if __name__ == "__main__":
   root=Tkinter.Tk()
   app=SimpleAppTk(root)
   app.grid(row=0,column=0)
   root.mainloop()
```
