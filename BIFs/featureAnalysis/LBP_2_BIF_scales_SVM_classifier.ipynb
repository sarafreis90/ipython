```python
import nibabel as nib
import numpy as np
import scipy.ndimage
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib as mpl
from scipy import stats
from scipy import ndimage
import glob
import csv
import os
import re
import math
import sys
import plotly.plotly as py
from plotly.graph_objs import *
from pylab import *


import sklearn
from sklearn import svm
from sklearn import datasets
from sklearn.cross_validation import train_test_split
from sklearn.grid_search import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.svm import SVC

from sklearn.metrics import roc_curve, auc

%matplotlib inline
```

```python
'''
sigma_scales:

0.0002mm
0.0003mm
0.00045mm
0.000675mm
0.0010125mm
0.00151875mm

'''

y_labels = np.loadtxt('/Volumes/vph-prism5/Sara/SPIE16/Manuscript/LightLines/LightLines_GLCM_labels.csv', delimiter = ',', usecols=(2,), dtype = int)
#filenames
z_names = np.loadtxt('/Volumes/vph-prism5/Sara/SPIE16/Manuscript/LightLines/LightLines_GLCM_labels.csv', delimiter = ',', usecols=(0,), dtype = str)


X_LBP = []
f_LBP = open('/Volumes/vph-prism5/Sara/SPIE16/Manuscript/LightLines_6scales/LightLines_LBP_features_0.000675mm.csv')
f_LBP.readline()  # skip the header
LBPreader = csv.reader(f_LBP)
for row in LBPreader:
    X_LBP.append(row)

X_LBP_2 = []
f_LBP_2 = open('/Volumes/vph-prism5/Sara/SPIE16/Manuscript/LightLines_6scales/LightLines_LBP_features_0.0010125mm.csv')
f_LBP_2.readline()  # skip the header
LBPreader_2 = csv.reader(f_LBP_2)
for row_2 in LBPreader_2:
    X_LBP_2.append(row_2)

print len(X_LBP)

merged_features=[]
for i in range(len(X_LBP)):
    merged_features.append(X_LBP[i]+X_LBP_2[i])

    
LBP_header = np.arange(17).tolist()
csv_LBP_Header = LBP_header + LBP_header


file = open('/Volumes/vph-prism5/Sara/SPIE16/Manuscript/LightLines_6scales/LBP_features_2scales_exp4.csv', "wb")
writer = csv.writer(file)
writer.writerow(csv_LBP_Header)
writer.writerows(merged_features)
file.close()

#print len(merged_features)
#print merged_features[0:2]

X_array_features = np.array(merged_features)
print X_array_features.shape

n_sample = len(X_array_features)

np.random.seed(0)
order = np.random.permutation(n_sample)
X_features = X_array_features[order]
y_labels = y_labels[order].astype(np.float)
z_names = z_names[order]

X_train = X_features[:.7 * n_sample] ## select 70% of the data to train
y_train = y_labels[:.7 * n_sample]
z_train = z_names[:.7 * n_sample]
X_test = X_features[.7 * n_sample:]
y_test = y_labels[.7 * n_sample:]
z_test = z_names[.7 * n_sample:]
print "Test set:"
print z_test

print "Training set:"
print z_train

```

```python
C=[]
gamma=[]
#for i in range(21): C.append(10.0**(i-5))
#for i in range(17): gamma.append(10**(i-14))

for i in range(6): C.append(10.0**(i-1))
for i in range(5): gamma.append(10**(i-5))
    
print C
print gamma
###-----------------------------------------------------------###
#---------- Set the parameters by cross-validation -----------###
###-----------------------------------------------------------###

'''
tuned_parameters = [{'kernel': ['rbf'], 'gamma': [1e-3, 1e-4, 1e-5],
                     'C': [1, 10, 100, 1000]},
                    {'kernel': ['linear'], 'C': [1, 10, 100, 1000]}]
'''

tuned_parameters = [{'kernel': ['rbf'], 'gamma': gamma,
                     'C': C},
                    {'kernel': ['linear'], 'C': C}]

#scores = ['precision', 'recall']
scores = ['accuracy', 'f1'] # optimize for accuracy rather than precision or recall, or for F1 if you have a situation with unbalanced classes.

```

```python
print "Cross-validation for multi-scale LightLines:"

for score in scores:
    print("# Tuning hyper-parameters for %s" % score)
    print()

    clf = GridSearchCV(SVC(C=1), tuned_parameters, cv=5,
                       scoring='%s' % score)
    clf.fit(X_train, y_train)

    print("Best parameters set found on development set:")
    print()
    print(clf.best_params_)
    print()
    print("Grid scores on development set:")
    print()
    for params, mean_score, scores in clf.grid_scores_:
        print("%0.3f (+/-%0.03f) for %r"
              % (mean_score, scores.std() * 2, params))
    print()

    print("Detailed classification report (Multi-scale LightLines):")
    print()
    print("The model is trained on the full development set.")
    print("The scores are computed on the full evaluation set.")
    print()
    y_true, y_pred = y_test, clf.predict(X_test)
    print(classification_report(y_true, y_pred))
    print()
    print("Confusion matrix")
    print(confusion_matrix(y_test, y_pred))
    print()
```

```python
##--------------------------------------------------------##
##-----------------------ROC curves-----------------------##
##--------------------------------------------------------##

print "True binary classification:"
print y_test
##----------------------------------##
##--------- LighLines LBP ----------##
##----------------------------------##

clf_LBP = svm.SVC(C=10000, kernel='rbf', gamma=0.1)
#clf_LBP = svm.SVC(C=10000, kernel='linear')
y_score = clf_LBP.fit(X_train, y_train).decision_function(X_test)

print "Predicted LBP:"
print clf_LBP.predict(X_test)

# Compute ROC curve and ROC area 

fpr_LBP, tpr_LBP, _ = roc_curve(y_test, y_score)
roc_auc_LBP = auc(fpr_LBP, tpr_LBP)


# Plot of a ROC curve for a specific class
plt.plot(fpr_LBP, tpr_LBP, label='LBP (BIF) (AUC = %0.2f)' % roc_auc_LBP)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
#plt.title('Receiver operating characteristic')
plt.legend(loc="lower right",prop={'size':9.5})
plt.show()
```
