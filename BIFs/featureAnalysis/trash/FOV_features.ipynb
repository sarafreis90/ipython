```python
###--------------------------------------------------------###
###------------- First order statistics -------------------###
###--------------------------------------------------------###
FOV_features = []

#for files in glob.glob('/Volumes/vph-prism5/Sara/TextureStuff/BIFs/KCL_ROIs/orientationAnalysis/experiment_02/results/lightLines_*_i.nii.gz'):
for files in glob.glob('/Volumes/vph-prism5/Sara/SPIE16/resultsLightLines/lightLines*.nii.gz'): 
    (dirname, filename) = os.path.split(files)
    filename = str(os.path.basename(files))
    filename = filename.replace('.nii.gz','')
    print filename

    img = nib.load(files)
    data = img.get_data() 
    data = data.squeeze()
    data = ndimage.rotate(data, 90, order=0)
    image = np.flipud(data)
    
    #print image.shape
    #print image.mean()
    #print image.std()
    #print stats.skew(image)
    #print stats.kurtosis(image)
    
    ### image is a 2-d array, need to change to 1-d to make a histogram
    nx, ny = image.shape # find the size of the array
    imgh = reshape(image, nx*ny) # change the shape to be 1d
    #print imgh.shape
    
    image_mean = imgh.mean()
    FOV_features.append(imgh.mean())
    
    image_std = imgh.std()
    FOV_features.append(imgh.std())
    
    image_skew = stats.skew(imgh)
    FOV_features.append(stats.skew(imgh))
    
    image_kurtosis = stats.kurtosis(imgh)
    FOV_features.append(stats.kurtosis(imgh))

print FOV_features
csvFOVfeaturesHeader = ['Mean', 'STD', 'Skewness', 'Kurtosis']    
#print csvFOVfeaturesHeader
#print FOV_features

FOV_features_all = chunks(FOV_features,4) 
#FOV_features_all = zip(*FOV_features_all)
print FOV_features_all
FOV_all_Filename = 'FOVFeatures.csv'
csvFilename_all = os.path.join(dirname, FOV_all_Filename )
file = open(csvFilename_all, "wb")
writer = csv.writer(file)
writer.writerow(csvFOVfeaturesHeader)
writer.writerows(FOV_features_all)
file.close()
```
