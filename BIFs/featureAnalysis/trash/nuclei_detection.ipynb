```python
##----- mask rgb images ------##
masked_image = np.zeros_like(img)
print 'Final masked Image:'
print masked_image.shape
print masked_image.dtype

# create mask with same dimensions as image
mask_3D = np.zeros_like(img)

# copy your image_mask to all dimensions (i.e. colors) of your image
for i in range(3): 
    mask_3D[:,:,i] = mask.copy()
print mask_3D.shape

# apply the mask to your image
idx = (mask_3D != 0)
masked_image[idx] = img[idx]

print masked_image.shape
        
plt.imshow(masked_image)
plt.axis('off')
plt.show()
```

```python
image_gray = rgb2gray(masked_image)

plt.imshow(image_gray, 'gray')
plt.axis('off')
plt.show()


blobs_log = blob_log(image_gray, max_sigma=30, num_sigma=10, threshold=.1)
# Compute radii in the 3rd column.
blobs_log[:, 2] = blobs_log[:, 2] * sqrt(2)

blobs_dog = blob_dog(image_gray, max_sigma=30, threshold=.1)
blobs_dog[:, 2] = blobs_dog[:, 2] * sqrt(2)

blobs_doh = blob_doh(image_gray, max_sigma=30, threshold=.01)

blobs_list = [blobs_log, blobs_dog, blobs_doh]
colors = ['yellow', 'lime', 'red']
titles = ['Laplacian of Gaussian', 'Difference of Gaussian',
          'Determinant of Hessian']
sequence = zip(blobs_list, colors, titles)

for blobs, color, title in sequence:
    fig, ax = plt.subplots(1, 1)
    ax.set_title(title)
    ax.imshow(masked_image, interpolation='nearest')
    for blob in blobs:
        y, x, r = blob
        c = plt.Circle((x, y), r, color=color, linewidth=2, fill=False)
        ax.add_patch(c)

    plt.axis('off')
plt.show()
```
