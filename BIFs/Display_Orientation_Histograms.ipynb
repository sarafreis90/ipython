```python
fig, ax = plt.subplots()

histo_list = [] # list of list: saves raw data from each csv file in a list
label_list = [] # list of list: saves the filenames
hist_values = [] # list of list: saves the final histogram results to be plotted
hist_values_norm = [] # list of list: saves the final histogram results NORMALISED (0 to 1) to be plotted

#for files in glob.glob('/Volumes/vph-prism5/Sara/TextureStuff/BIFs/RUNMC001/normalROIs/BIFs_results_wholeImage/maskedResults/orientation_I10_02_normal_0.002mm_*.nii.gz'):
for files in glob.glob('/Volumes/vph-prism5/Sara/TextureStuff/BIFs/KCL_ROIs/orientationAnalysis/experiment_02/results/orientation_23_03_m.nii.gz'):
    
    (dirname, filename) = os.path.split(files)
    filename = str(os.path.basename(files))
    print filename

    img = nib.load(files)
    data = img.get_data() 
    data = data.squeeze()
    data = ndimage.rotate(data, 90, order=0)
    data = np.flipud(data)
    histo_list.append(data)
    
    plt.imshow(data)
    plt.axis('off')
    plt.show()
    
    try:
        histo_patho = re.search('_(.+?).nii.gz', filename).group(1) # was orientation_I6_02_normal_0.001mm_1.nii.gz, becomes I6_02_normal_0.001mm_1
        label_list.append(histo_patho)
        
    except AttributeError:
        # AAA, ZZZ not found in the original string
        histo_patho = '' # apply your error handling
        

print "raw data:"
#print histo_list
print "length of list:"
print len(histo_list)
print label_list
print histo_patho

xmin = 0
xmax = pi/2
step = (pi/2)/4 # 4 = 30 degrees; 3 = 45 degrees; 7 = 15 degrees; 18 = 5 degrees
bins = np.linspace(xmin, xmax, (xmax-xmin)/step)
    
for i in range(len(histo_list)):    
    hist, bin_edges = np.histogram(histo_list[i][histo_list[i] != 0], bins=np.linspace(xmin, xmax, (xmax-xmin)/step))
    print"histogram:"
    print hist
    print bin_edges
    hist_values.append(hist)
print hist_values
    
# normalise the histogram
for i,j in enumerate(hist_values): #i=position, j=value
    hist_values_norm.append(j/sum(j))

# Plot colormap
cmap = mpl.cm.hsv

bar_width = ((bin_edges[1]-bin_edges[0])/len(histo_list))
print len(histo_list)
#print bar_width

fig, ax = plt.subplots()
for j, histogram in enumerate(hist_values_norm):
    
    p = ax.bar(bin_edges[:-1]+(j*bar_width), histogram, width=((bin_edges[1]-bin_edges[0])/len(histo_list)), color=cmap(j/float(10)), label=label_list[j])

    ax.set_xlabel('Angles')
    ax.set_ylabel('Occurences (%)')
    ax.set_ylim(bottom=0)
    ax.set_ylim(top=0.6)
#ax.set_title('')
    ax.set_xticks(bin_edges[:-1]+(j*bar_width))
    #ax.set_xticklabels( ('0-15', '15-30', '30-45', '45-60', '60-75', '75-90') )
    ax.set_xticklabels( ('0-30', '30-60', '60-90') )
    ax.legend(loc=1, borderaxespad=0., fontsize='x-small', fancybox=True, framealpha=0.5)    

plt.grid(True)
plt.show  
#plt.savefig('/Volumes/vph-prism5/Sara/TextureStuff/BIFs/RUNMC001/normalROIs/histoAnalises/rim_histogram_I10_02_normal_0.002.png') 
#plt.savefig('/Volumes/vph-prism5/Sara/TextureStuff/BIFs/ROIsFromKCL/histoAnalises/histogram_TS64_0.002mm.png')

```
