```python
import nibabel as nib
import numpy as np
import scipy.ndimage
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib as mpl
from scipy import stats
import glob
import csv
import os
import re
import math
from math import pi
from __future__ import division
from scipy import misc
import cv2
from scipy import ndimage
from nibabel import nifti1
%matplotlib inline
```

```python
##---------------------------------------------------------##
##--------- Read LightLines + orientation Image -----------##
##---------------------------------------------------------##

for lightLines in glob.glob('/Volumes/vph-prism5/Sara/forJohn/wholeImageAnalysis/BIFs_results_wholeImage/lightLines_I10_02_normal_0.002mm.nii.gz'):
    
    lightLines_img = nib.load(lightLines)
    lightLines_data = lightLines_img.get_data()
    lightLines_data = lightLines_data.squeeze()
    lightLines_data = ndimage.rotate(lightLines_data,90)
    lightLines_data = np.flipud(lightLines_data)
    print lightLines_data.shape
    
    plt.imshow(lightLines_data, 'gray')
    plt.axis('off')
    plt.show()

```

```python
for orientation in glob.glob('/Volumes/vph-prism5/Sara/forJohn/wholeImageAnalysis/BIFs_results_wholeImage/orientation_I10_02_normal_0.002mm.nii.gz'):
        
        orientation_img = nib.load(orientation)
        orientation_data = orientation_img.get_data()
        orientation_data = orientation_data.squeeze()
        orientation_data = ndimage.rotate(orientation_data,90)
        orientation_data = np.flipud(orientation_data)
        print lightLines_data.shape
    
        plt.imshow(orientation_data, 'gray')
        plt.axis('off')
        plt.show()
```

```python
new_orientation = np.zeros(orientation_data.shape).astype(np.float32) 

for lines in range(lightLines_data.shape[0]):
    for col in range(lightLines_data.shape[1]): 
        if (lightLines_data[lines,col]>0.0) and (lightLines_data[lines,col]<7.0):
            new_orientation[lines,col] = orientation_data[lines,col]
    #if ((pixel > 0) and (pixel < 5.0)):
     #   new_orientation[pixel]=rientation_data[pixel]

#new_orientation[pixel]=orientation_data[pixel]
#plt.imshow(new_orientation, 'gray')
#plt.axis('off')
#plt.show()
```

```python
plt.imshow(new_orientation, 'gray')
plt.axis('off')
plt.show()

save_new_orientation = nib.Nifti1Image(new_orientation, np.eye(4))
save_new_orientation.get_data_dtype() == np.dtype(np.float32)

nib.save(save_new_orientation,'/Volumes/vph-prism5/Sara/forJohn/wholeImageAnalysis/BIFs_results_wholeImage/thresholded_orientation_I10_02_normal_0.002mm.nii.gz')
```
