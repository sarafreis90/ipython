```python
##-------------------------------------------------------------------------------##
##--------- Read mask generated by itk-snap/weka and perform dilation -----------##
##-------------------------------------------------------------------------------##

def disk_structure(n):
    struct = np.zeros((2 * n + 1, 2 * n + 1))
    x, y = np.indices((2 * n + 1, 2 * n + 1))
    mask = (x - n)**2 + (y - n)**2 <= n**2
    struct[mask] = 1
    return struct.astype(np.int)

original_mask = []
binary_images = []
rim_images = []
final_filename = []

##-------load binary image (either from ITK-SNAP or weka classication)-------##
for files in glob.glob('/Volumes/vph-prism5/Sara/TextureStuff/BIFs/RUNMC001/normalROIs/snapMasks/mask_I8_01_normal.nii.gz'):
    
    (dirname, filename) = os.path.split(files)
    new_filename = re.sub('mask_','rim_',filename)
    new_filename = re.sub('.nii.gz','',new_filename)
    print files
    print new_filename
    
    img = nib.load(files)
    data = img.get_data()
    data = data.squeeze()
    data = ndimage.rotate(data,90)
    data = np.flipud(data)
    print data.shape
    original_mask.append(data)
    
    plt.imshow(data, 'gray')
    plt.axis('off')
    plt.show()

##----COMPUTE DILATION----#
    for i in xrange(3):
        
        t = time.clock()
        
        #selem = disk(100*i)
        #dilated_image = binary_dilation(data,selem)
        dilated_image = ndimage.binary_dilation(data, structure=disk_structure(100), iterations=(2*(i+1))).astype(np.int)
        #dilated_image = ndimage.binary_dilation(data, structure=np.ones((100,100))).astype(np.int)

        tt = time.clock()
        print tt-t
        
        binary_images.append(dilated_image)
    
        plt.subplot(1,len(xrange(3)),i+1),plt.imshow(dilated_image,'gray')
        plt.title('rim')
        plt.axis('off')
        plt.subplots_adjust(wspace=0, hspace=0., top=0.99, bottom=0.01,
                    left=0.01, right=0.99)
    
    plt.show()
    binary_images = original_mask + binary_images
```

```python
##----------------------------------------------------------------------##
##--------- Mask LightLines + Orientation image with rim mask ----------##
##----------------------------------------------------------------------##
def rgb2gray(rgb):

    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b

    return gray

orientation_images = []
masks_img = []
masked_img = []

for images in glob.glob('/Volumes/vph-prism5/Sara/TextureStuff/BIFs/RUNMC001/normalROIs/BIFs_results_wholeImage/orientation_I10_02_normal_0.002mm.nii.gz'):
    
    (dirname_res, filename_res) = os.path.split(images)
    new_folder = 'maskedResults'
    masked_dirname = os.path.join(dirname_res, new_folder)
    print masked_dirname
    masked_filename = re.sub('.nii.gz','', filename_res)
    
    img = nib.load(images)
    orientation_img = img.get_data()
    orientation_img = ndimage.rotate(orientation_img,90)
    orientation_img = np.flipud(orientation_img)
    
    print 'Original Image:'
    print orientation_img.shape
    print orientation_img.dtype
    
    orientation_images.append(orientation_img)
        
    plt.subplot(121),plt.imshow(orientation_img)
    plt.axis('off')
    plt.show()
    
    k = 1
    for masks in glob.glob('/Volumes/vph-prism5/Sara/TextureStuff/BIFs/RUNMC001/normalROIs/rimMasks/rim_I10_02_normal_*.tif'):
        print masks
        
        mask = cv2.imread(masks)
        mask = rgb2gray(mask).astype(np.float32)  # convert image from (4200, 3300, 3) to (4200, 3300)
        masks_img.append(mask)
        
        print 'Mask Image:'
        print mask.shape
        print mask.dtype
        
        masked_image = np.zeros(orientation_img.shape).astype(np.float32) 
        print 'Final masked Image:'
        print masked_image.shape
        print masked_image.dtype
        
        idx = (mask != 0)
        print idx
        masked_image[idx]=orientation_img[idx]
        
        masked_img.append(masked_image)
        
        
        number = str(k)
        new_filename = masked_filename + '_' + number + '.nii.gz'
        print new_filename

        
        #SAVE MASKED IMAGES
        save_masked = nib.Nifti1Image(masked_image, np.eye(4))
        save_masked.get_data_dtype() == np.dtype(np.float32)
        #img.header.get_xyzt_units()

        nib.save(save_masked, os.path.join(masked_dirname, new_filename) )

        
        for i in xrange(len(masked_img)):
            plt.subplot(1,len(masked_img),i+1),plt.imshow(masked_img[i], 'gray')
            plt.axis('off')
        k = k+1
    
    plt.show() 
        
```
