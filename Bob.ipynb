```python
import nibabel as nib
import numpy as np
import scipy.ndimage
import matplotlib.pyplot as plt
import matplotlib as mpl
from scipy import stats
from matplotlib.pyplot import *

barWidth = 1.0

deltaZVals = np.array( [2.0,3.0,5.0,6.0,10.0] ) * 1000.0 * 0.004
meanAbsDistVals = [5, 8, 9, 10, 11]
relOverSegs = np.arange(5)
relUnderSegs = np.arange(5)

f=plt.gcf()
f.clear()
axes = f.add_subplot(1,1,1, axisbg='white')

axes.bar(deltaZVals-barWidth/2, relOverSegs, width=barWidth, color='green', label="Over Segmentation")
axes.bar(deltaZVals+barWidth/2, relUnderSegs, width=barWidth, color='red', label="Under Segmentation")

axes.legend(loc=9, borderaxespad=0., fontsize='x-small', fancybox=True, framealpha=0.5)
#plt.grid(color='gray', linestyle='-')

ax=plt.gca()   

#ax.yaxis.grid( True, color='white', linestyle='-' )
ax.spines['bottom'].set_color('black')
ax.spines['top'].set_color('black')
ax.spines['right'].set_color('black')
ax.spines['left'].set_color('black')
ax.xaxis.label.set_color('black')
ax.yaxis.label.set_color('black')
ax.tick_params(axis='x', colors='black')
ax.tick_params(axis='y', colors='black')
ax.set_ylim(bottom=0)
ax.set_ylim(top=35)

#ax.set_xlabel('$\\Delta z \\; [\\unit{\\micro\\metre}]$')
#ax.set_ylabel('$V_{o/u}/V_H \\; [ \\% ]$')

newax = ax.twinx()
#newax.bar(deltaZVals+barWidth/2, meanAbsDistVals, width=barWidth/2, color = 'white', label="Abs. Distance")
newax.errorbar(deltaZVals, meanAbsDistVals, ecolor = 'black', color='black', label="Abs. Distance")
newax.legend(loc=0, borderaxespad=0., fontsize='x-small', fancybox=True, framealpha=0.5)
#newax.set_ylabel('$\\bar{d} \\; [\\unit{\\micro\\metre}]$')
newax.set_ylim(bottom=0)
newax.set_ylim(top=60)
newax=plt.gca()

newax.yaxis.label.set_color('black')
newax.tick_params(axis='y', colors='black')

#newax.yaxis.label.set_color('black')
#newax.tick_params(axis='y', colors='black')

plt.show()
#plt.savefig( "/Users/sarareis/Dropbox/PhD/SPIE/Presentation/test.pdf", bbox_inches='tight' )
```
