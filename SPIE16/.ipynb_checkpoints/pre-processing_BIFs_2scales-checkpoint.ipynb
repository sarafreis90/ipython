```python
###---------------------------------------------------------###
###-----Extract local pixel histogram from BIFs images------###
###---------------------------------------------------------###

def create_feat_vector(image, maskI):
    
    filename_list = []
    bifs_feature_vector = []

    for files in sorted(glob.glob(image)):    
        (dirname, filename) = os.path.split(files)
        filename = str(os.path.basename(files))
        if (re.search('_0.0004mm.nii.gz', filename)):
            filename = filename.replace('_0.0004mm.nii.gz','')
            print filename
            filename_list.append(filename)
        else:
            filename = filename.replace('_0.0006mm.nii.gz','')
            print filename
            filename_list.append(filename)

        img = nib.load(files)
        data_i = img.get_data() 
        data_i = data_i.squeeze()
        data_i = ndimage.rotate(data_i, 90, order=0)
        image_i = np.flipud(data_i)
        image_i = np.uint8(image_i)
        
        '''
        print image_i.dtype
        print image_i.shape
        print np.amin(image_i)
        print np.amax(image_i)

        plt.imshow(image_i, 'gray')
        plt.show()
        '''

        ##----- Load correspondent binary mask -----##
        mask_name = 'mask_' + filename + '.nii.gz'
        mask_in = os.path.join(maskI, mask_name )

        mask_img = nib.load(mask_in)
        mask = mask_img.get_data()
        mask = mask.squeeze()
        mask = ndimage.rotate(mask,90)
        mask = np.flipud(mask)
        mask = mask.astype(uint8)
        #print mask.shape
        #print mask.dtype

        #print np.amax(masked_image)
        '''
        hist,bins = np.histogram(image_i.ravel(), 6, [0,6])
        plt.hist(image_i.ravel(), 6, [0,6])
        plt.show()
        '''

        hist_full = cv2.calcHist([image_i],[0],None,[7],[0,7])
        hist_mask = cv2.calcHist([image_i],[0],mask,[7],[0,7])
        #print hist_mask.shape
        #print hist_mask.dtype
        hist_mask_r = hist_mask.reshape(1,7).tolist()

        bifs_feature_vector = bifs_feature_vector + hist_mask_r


        '''
        plt.subplot(221), plt.imshow(image_i, 'gray')
        plt.subplot(222), plt.imshow(mask,'gray')
        plt.subplot(223), plt.plot(hist_full), plt.plot(hist_mask)
        plt.show()
        '''

    return bifs_feature_vector

link1 = '/Volumes/vph-prism5/Sara/SPIE16/BIFs_experiments/*0.0004mm.nii.gz'
link2 = '/Volumes/vph-prism5/Sara/SPIE16/BIFs_experiments/*0.0006mm.nii.gz'
mask12 = '/Volumes/vph-prism5/Sara/SPIE16/Experiments_v2.0/stroma_masks'

scale1 = create_feat_vector(link1, mask12)
scale2 = create_feat_vector(link2, mask12)

    
```

```python
###---------------------------------------------------------###
###-----Merge the 2 BIFs scales in one feature vector-------###
###---------------------------------------------------------###

'''
Multi-scale sigma scales:

0.0004mm
0.0006mm

'''

E1=[]
for i in range(len(scale1)):
    E1.append(scale1[i]+scale2[i])

E1_LBP_header = np.arange(len(scale1[0])).tolist()
merged_E1_LBP_Header = E1_LBP_header + E1_LBP_header
print "Header:"
print merged_E1_LBP_Header

file = open('/Volumes/vph-prism5/Sara/SPIE16/BIFs_experiments/BIFs_2scales.csv', "wb")
writer = csv.writer(file)
writer.writerow(merged_E1_LBP_Header)
writer.writerows(E1)
file.close()
```
